投稿数が多いもの、もしくは、補足？が多いもの、もしくは、voteが多いもの

```python
SITE = StackAPI('ja.stackoverflow')
```

```python
comments = SITE.fetch('questions', min=5, tagged='python', sort='votes')
print("votesが5以上の投稿")
for i in range(len(comments["items"])):
    #print(i)
    if "link" in comments["items"][i]["owner"]:
        url = comments["items"][i]["link"]
        print(url)

#comments["items"][3]
```
